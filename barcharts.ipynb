# Bar Charts

```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
# Convert data to dataframe
lung_df = pd.read_csv(os.path.join("Resources", "CDC_Cancer.csv"))
lung_df.head()
```

```python
# Get rid of unnecessary columns
lung_df = lung_df.drop(["Notes", "State Code", "Year Code", "Sex Code", "Race Code", "Age Group"], axis=1)
lung_df = lung_df.rename(columns = {"Age Group Code":"Age Group"})
lung_df.head()
```

```python
#create df
states_lungdf = lung_df[lung_df.Year == 2015]
states_lungdf=lung_df.groupby(["State"]).agg({"Deaths":"sum"})
states_lungdf
```

```python
states_lungdf.Deaths.sort_values()
#State with the highest count of lung cancer deaths= California -13,177 within 2015 
#State with the lowest count lung cancer deaths= District of Columbia- 491 deaths within 2015
```

```python
barlabels = list(states_lungdf.Deaths)
```

```python
#set x axis and tick locations
x_axis = np.arange(len(states_lungdf))
tick_locations = [value for value in x_axis]

#plot
states_lungdf.plot(kind="bar", figsize=(20,3))

#set x and y limits
plt.xlim(-0.75, len(x_axis))
barlabels.

plt.title("Lung Cancer Mortality across the U.S. within 2015")
plt.xlabel("States")
plt.show("Total of Lung Cancer Deaths")
plt.tight_layout()
```
