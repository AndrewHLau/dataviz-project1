```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
# Convert data to dataframe
gun_df = pd.read_csv(os.path.join("Resources", "gun_data.csv"))
gun_df.head()
```

```python
# Get list of years and races
years = gun_df["Year"].unique()
races = gun_df["Race"].unique()
```

```python
# Group by year and by race and sum the deaths

race_df = gun_df.groupby(["Year", "Race"])["Deaths"].sum().unstack()
race_df.head()
```

```python
# Set parameters for plot
plt.figure(figsize =(8,4))
plt.title("Gun Deaths by Race", fontsize = 20)
plt.xlabel("Year", fontsize = 16)
plt.ylabel("Number of deaths", fontsize = 16)
plt.grid()

# Function to create scatter plot for gun mortality
def gun_plot(race):
    y = race_df[race]
    plt.plot(years, y)
    return
gun_plot("White")

plt.legend()
# Display plot
plt.show()
```

```python
# Set parameters for plot
plt.figure(figsize =(12,8))
plt.title("Gun Deaths by Race", fontsize = 20)
plt.xlabel("Year", fontsize = 16)
plt.ylabel("Number of deaths", fontsize = 16)
plt.grid()

# Function to create scatter plot for gun mortality
def gun_plot(race):
    y = race_df[race]
    plt.plot(years, y)
    return

for race in races:
    gun_plot(race)

plt.legend()
# Display plot
plt.show()
```
